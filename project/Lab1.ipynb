Sampling

```python
import numpy as np  # Import the numpy library for numerical operations
import matplotlib.pyplot as plt  # Import the matplotlib library for plotting

# Define parameters
Ts = 0.1  # Sampling interval (time step) in seconds
N = 100  # Number of samples (points) in the signal
w0 = 2  # Angular frequency of the sine wave in radians per second

# Generate the time vector
t = np.arange(N) * Ts  # Create an array of time points from 0 to (N-1)*Ts

# Generate the sine wave signal
x = np.sin(t * w0)  # Compute the sine of each phase value (t * w0)

# Plot the signal
plt.plot(t, x)  # Plot the sine wave (x) against the time vector (t)
plt.grid()  # Add a grid to the plot for better readability
plt.xlabel('time - secs')  # Label the x-axis as "time - secs"
plt.ylabel('signal x')  # Label the y-axis as "signal x"
plt.title('Plot of x vs t')  # Set the title of the plot to "Plot of x vs t"
plt.show()  # Display the plot

```

Plotting a signal

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(100) * 0.1
x = np.sin(t * 2)
y = np.cos(t * 2)

plt.plot(t, x, '-', label='x')
plt.plot(t, y, '--', label='y')
plt.grid()
plt.xlabel('time - secs')
plt.ylabel('signal')
plt.title('Plot of x and y vs t')
plt.legend()
plt.show()
```

Subplot

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(100) * 0.1
x = np.sin(t * 2)
y = np.cos(t * 2)

plt.subplot(2, 1, 1)
plt.plot(t, x, 'g')
plt.grid()
plt.xlabel('time - secs')
plt.ylabel('signal x')
plt.title('Plots of x and y')

plt.subplot(2, 1, 2)
plt.plot(t, y, 'r')
plt.grid()
plt.xlabel('time - secs')
plt.ylabel('signal y')

plt.show()
```

Figure command

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(100) * 0.1
x = np.sin(t * 2)
y = np.cos(t * 2)

plt.figure(1)
plt.plot(t, x)
plt.grid()
plt.xlabel('time - secs')
plt.ylabel('signal x')

plt.figure(2)
plt.plot(t, y)
plt.grid()
plt.xlabel('time - secs')
plt.ylabel('signal y')

plt.show()
```

Listening to the signal

```python
import numpy as np
from scipy.io.wavfile import write

Fs = 8000  # Sample rate
t = np.linspace(0, 1, Fs)
x = np.sin(2 * np.pi * 220 * t)

write('sound.wav', Fs, x.astype(np.float32))
```

Writing an M-file

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write

T = 10  # signal duration
Fs = 8000  # sampling frequency
t = np.arange(0, T, 1/Fs)  # time axis
f = 1  # frequency
x = np.exp(-2 * t) * np.sin(2 * np.pi * f * t)

plt.plot(t[:300], x[:300], 'r')
plt.grid()
plt.xlabel('time - secs')
plt.ylabel('signal value - volts')
plt.title('A Plot of a Simple Signal')
plt.show()

write('simple_signal.wav', Fs, x.astype(np.float32))
```

Plotting Pure signals

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write

Fs = 8000  # sampling frequency in Hz
f = 261.63  # tone frequency Hz (Middle C)
w = 2 * np.pi * f  # tone frequency rad/sec
T = 3  # duration of signal in seconds
t = np.arange(0, T, 1/Fs)  # vector of sample times
Amp = 1  # amplitude of tone
ph = 0  # phase of tone
x = Amp * np.sin(w * t + ph)  # the vector of signal values
N = 300  # no. of samples to be plotted

plt.plot(t[:N], x[:N])
plt.grid()
plt.xlabel('time - secs')
plt.ylabel('signal value')
plt.title('Pure Tone of Middle C')
plt.show()

write('pure_tone.wav', Fs, x.astype(np.float32))
```
